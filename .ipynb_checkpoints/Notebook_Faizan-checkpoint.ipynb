```python
import numpy as np 
import pandas as pd 
import os
import json
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
import random
import warnings
warnings.filterwarnings("ignore")
import scipy as sp
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
print("Shape of train is {}".format(train.shape))
print("Shape of test is {}".format(test.shape))
```

```python
print(train.head())
```

```python
print(train.info())
```

```python
# Insight_1: There are 14993 (around 15k pets in the dataset)
# Insight_2: There are less names and descriptions (13736 and 14981 respectively) than the number of pets which implies that not all pets have a name/description
```

```python
plt.figure(figsize=(10,6))
ax = sns.countplot(x="AdoptionSpeed", data=train)
plt.title('Frequency of Adoptions wrt Adoption Speed')
plt.xlabel('Adoption Speed')
plt.ylabel('Frequency [%]')

total = len(train['AdoptionSpeed'])
        
for p in ax.patches:
        ax.annotate('{:.1f}%'.format(100 * p.get_height()/total), (p.get_x()+0.3, p.get_height()+50))
        
plt.savefig('Frequency of Adoptions wrt Adoption Speed.png')
```

```python
# Insight_3: For majority of the pets, there was no adoption after 100 days of being listed.
```

```python
plt.figure(figsize=(10,6))
ax = sns.countplot(x="Type", data=train)
plt.title('Frequency of each type of pet')
plt.xlabel('Type of Pet')
plt.ylabel('Frequency [%]')
plt.xticks([0, 1], ['Dog', 'Cat'])

total = len(train['AdoptionSpeed'])
        
for p in ax.patches:
        ax.annotate('{:.1f}%'.format(100 * p.get_height()/total), (p.get_x()+0.3, p.get_height()+50))
        
plt.savefig('Frequency of each type of pet.png')
```

```python
# Insight_4: There are more number of dogs than cats in the dataset
```

```python
plt.figure(figsize=(10,6))
ax = sns.countplot(x="Vaccinated", hue="AdoptionSpeed",data=train)
plt.title('AdoptionSpeed and Vaccinated')
plt.xlabel('Vaccinated')
plt.ylabel('Frequency [%]')
plt.xticks([0, 1, 2], ['Yes', 'No', 'Not sure'])

total = len(train['AdoptionSpeed'])
        
for p in ax.patches:
        ax.annotate('{:.1f}%'.format(100 * p.get_height()/total), (p.get_x()+0.01, p.get_height()+50))

plt.savefig('AdoptionSpeed and Vaccinated.png')
```

```python
plt.figure(figsize=(10,6))
ax = sns.countplot(x="Dewormed", hue="AdoptionSpeed",data=train)
plt.title('AdoptionSpeed and Dewormed')
plt.xlabel('Dewormed')
plt.ylabel('Frequency [%]')
plt.xticks([0, 1, 2], ['Yes', 'No', 'Not sure'])

total = len(train['AdoptionSpeed'])
        
for p in ax.patches:
        ax.annotate('{:.1f}%'.format(100 * p.get_height()/total), (p.get_x()+0.01, p.get_height()+50))
        
plt.savefig('AdoptionSpeed and Dewormed.png')
```

```python
plt.figure(figsize=(10,6))
ax = sns.countplot(x="Sterilized", hue="AdoptionSpeed",data=train)
plt.title('AdoptionSpeed and Sterilized')
plt.xlabel('Sterilized')
plt.ylabel('Frequency [%]')
plt.xticks([0, 1, 2], ['Yes', 'No', 'Not sure'])

total = len(train['AdoptionSpeed'])
        
for p in ax.patches:
        ax.annotate('{:.1f}%'.format(100 * p.get_height()/total), (p.get_x()+0.01, p.get_height()+50))
        
plt.savefig('AdoptionSpeed and Sterilized.png')
```

```python
plt.figure(figsize=(10,6))
ax = sns.countplot(x="Health", hue="AdoptionSpeed",data=train)
plt.title('AdoptionSpeed and Health')
plt.xlabel('Health')
plt.ylabel('Frequency [%]')
plt.xticks([0, 1, 2], ['Healthy', 'Minor Injury', 'Serious Injury'])

total = len(train['AdoptionSpeed'])
        
for p in ax.patches:
        ax.annotate('{:.1f}%'.format(100 * p.get_height()/total), (p.get_x()+0.01, p.get_height()+50))
        
plt.savefig('AdoptionSpeed and Health.png')
```

Insights for Vaccinated, Dewormed, Sterilized and Health features:
Majority of the animals are healthy as previously discovered. There are very few animals having minor injury. These animals also face less adoption rates. Same is the case with animals having serious injury. There is an adoption preference for:
    1. Non-vaccinated animals
    2. Non-sterilized animals
    3. Dewormed animals
    4. Healthy animals
