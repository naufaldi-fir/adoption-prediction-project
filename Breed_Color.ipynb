```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df=pd.read_csv('train.csv')
breed=pd.read_csv('breed_labels.csv')
color=pd.read_csv('color_labels.csv')

df_color=df[["Color1", "Color2", "Color3"]]
df_breed=df[["Breed1", "Breed2"]]

print("Breed:")
breed1_unique=df_breed['Breed1'].unique()
breed2_unique=df_breed['Breed2'].unique()
print(sorted(breed1_unique))
print(sorted(breed2_unique))
print(breed[:4])
print(breed[303:])
print(breed[238:242])
print(breed.describe())

print("Color:")
color1_unique=df_color['Color1'].unique()
color2_unique=df_color['Color2'].unique()
color3_unique=df_color['Color3'].unique()
print(sorted(color1_unique))
print(sorted(color2_unique))
print(sorted(color3_unique))
print(color)

#Insights Breed: In the BreedID file, there are 307 different names for the breeds. The entry with index 240 is not a certain breed, but is called "Mixed Breed". At first I thought that this should not have been a possible option for breed, but now I think that this simply means that one of the parents of the dogs was of "mixed breed", which would not result in any problem for the dataset.

#Insights Color: There are 7 different colors.
```

```python
pie_group = df_breed.groupby("Breed1")["Breed1"].count()
print(pie_group)
pie_group.plot.pie()
print(breed[264:267])

#Insights: A lot of pets have at least one parent, that is breed 307, which is "mixed breed". Breed ID 265 and 266 also occur often, these are "domestic medium hair" and "domestic short hair". 
#Possible Problem: Domestic Medium/Short/Long hair might just be a general term for a certain group of breeds, which means that many of the other breeds are actually a type of this Domestic Medium/Short/Hair breed. 
```
